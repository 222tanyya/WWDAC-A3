```python
import pandas as pd
```

```python
df = pd.read_csv("/content/wwdac NintendoGames CLEANED.csv")
df.head()
```

```python
import ast
df['genres'] = df['genres'].apply(ast.literal_eval)
df.head()
```

```python
genres_encoded = df['genres'].explode().str.get_dummies().groupby(level=0).sum()
```

```python
genres_encoded = genres_encoded.add_prefix('genre')
```

```python
nintendo_games_encoded = pd.concat([df, genres_encoded], axis=1)
```

```python
nintendo_games_encoded.head()
```

```python
output_path = "/content/wwdac_NintendoGames_CLEANED_ENCODED.csv"
nintendo_games_encoded.to_csv(output_path, index=False)
```
